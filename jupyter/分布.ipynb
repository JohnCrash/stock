```python
import numpy as np
import time
import stock
import kline
import choose
import math
import matplotlib.pyplot as plt
```

# boll 宽度和macd成功之间的关系

```python
def bollMacdR(code):
    c,k,d = stock.loadKline(code)
    bo = stock.bollLineK(k,20)
    bw = stock.bollWidth(bo)
    macd = choose.macdRange(k)
    t = choose.buySellMatch(macd[:,0],macd[:,1])
    r = choose.buySellReturnRate(k,t)
    w = bw[t[:,0]]
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    axs[0].hist(w,bins=20)
    axs[1].hist(r,bins=20)
    axs[2].hist2d(w,r)
```

```python
bollMacdR('SH:600460')
```

# 周macd值和macd成功之间的关系

```python
def weekMacdDayMacdR(code):
    c,k,d = stock.loadKline(code)
    wk = stock.mergeK(k,5)
    wmacd,_ = stock.macd(wk)
    wd = stock.scaleTo1d(wmacd,len(k))
    
    macd = choose.macdRange(k)
    t = choose.buySellMatch(macd[:,0],macd[:,1])
    r = choose.buySellReturnRate(k,t)
    w = wd[t[:,0]]
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    axs[0].hist(w,bins=20)
    axs[1].hist(r,bins=20)
    axs[2].hist2d(w,r)
```

```python
weekMacdDayMacdR('SH:600460')
```
